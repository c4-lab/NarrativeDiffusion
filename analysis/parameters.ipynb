# Parameter inspection

### Social influence

Looking at the properties of the sigmoid for social influence

```python


import numpy as np
import matplotlib.pyplot as plt

I_scale = .2
I_shift = 0

def I(n):
    return 2 / (1 + np.exp(-I_scale*(n-I_shift))) - 1

x = np.linspace(0, 30, 400)
y = I(x)

plt.plot(x, y, label=r'$\frac{2}{1 + e^{-n_i*s}} - 1$')
plt.xlabel('$n_i$ (Number of Neighbors Adopted)')
plt.ylabel('$I(s_i)$ (Social Influence)')
plt.title('Social Influence as a function of Adopted Neighbors')
plt.legend()
plt.grid(True)
plt.show()
```

### Narrative influence 

Looking at the properties of the sigmoid for narrative influence

```python
import numpy as np
import matplotlib.pyplot as plt

alpha = .9  # Slope of sigmoid
x_0 = 5 # fifty percent probability
gamma = 3 #scaling factor
beta = 2 #baseline


def p(w):
    return 1 / (1 + np.exp(-alpha*(beta + gamma * w - x_0)))

x = np.linspace(0, 8, 300)
y = p(x)

plt.plot(x, y)
plt.xlabel('$n_i$ (Number of Neighbors Adopted)')
plt.ylabel('$p_i(t)$ (Adoption probability)')
plt.title('Raw weight')
plt.grid(True)
plt.show()
```
